```python
df = pd.read_csv('final.csv', usecols=lambda column: column != "Unnamed: 0")
```

```python
df['year'] = df['publicacion'].apply(lambda x: x[-4:])
```

```python
df['Book_name'] = df['Book_name'].apply(lambda x: x.replace("'",""))
```

```python
def estructura_texto(row):
  texto = f"""Title: {row.Book_name}
Author: {row.writers}
Year: {row.year}
Description: {row.description}
rating: {row.rating}
  """
  return texto
```

```python
for row in df.itertuples():
        estructura_texto(row)
    # print(row) # se accede a cada columnas con punto
```

```python
df['textual_representation'] = df.apply(estructura_texto, axis=1)
```

```python
import numpy as np
import requests
import time

# Dimensiones
faiss_dim = 4096  # Dimensin para FAISS
embedding_dim = 768  # Dimensin para embeddings

# Inicializa FAISS y la matriz de embeddings
index = faiss.IndexFlatL2(faiss_dim)
num_rows = len(df['textual_representation'])
X = np.zeros((num_rows, embedding_dim), dtype='float')

for i, repre in enumerate(df['textual_representation']):
    if i % 100 == 0:
        print(f"Processing row {i}...")
    try:
        headers = {'Content-Type': 'application/json'}
        res = requests.post('http://127.0.0.1:11434/v1/embeddings',
                            json={'model': 'llama3.2:1b', 'prompt': repre}, headers=headers)


        res.raise_for_status()
        embedding = res.json().get('embedding')
        
        # Validate the response
        if embedding is None:
            raise ValueError("Embedding not found in response")
        
        # Store the embedding
        X[i] = np.array(embedding)
    except Exception as e:
        print(f"Error processing row {i}: {e}")
        continue
    time.sleep(0.1)  # Optional delay to avoid overloading the server
```

```python
import requests

# URL de la API de Ollama
url = "http://localhost:11434/v1/models"

# Realizar la solicitud GET para obtener los modelos disponibles
response = requests.get(url)

# Comprobar si la solicitud fue exitosa
if response.status_code == 200:
    print("Conexin exitosa. Modelos disponibles:", response.json())
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
import requests
import json

url = "http://localhost:11434/v1/embeddings"

# El texto para generar un embedding
text_input = "Ollama es una herramienta poderosa para procesamiento de lenguaje natural."

# Crear el payload para la solicitud
payload = {
    "text": text_input,
    "model": "llama3.2:1b"  # Asegrate de usar el modelo correcto
}

# Definir los encabezados de la solicitud
headers = {
    'Content-Type': 'application/json',
}

# Realizar la solicitud POST
response = requests.post(url, data=json.dumps(payload), headers=headers)

# Verificar la respuesta
if response.status_code == 200:
    embedding = response.json()
    print("Embedding generado:", embedding)
else:
    print(f"Error {response.status_code}: {response.text}")
```
